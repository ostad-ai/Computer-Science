# Euler method
The **Euler method** is the most basic numerical technique for approximating solutions to ordinary differential equations (ODEs). Consider we have an **initial value problem** (IVP) of the form:
<div style="margin-top:5pt"></div>
$\large \dot{y}=f(y,t)$ and $\large y(t_0)=y_0$
<div style="margin-bottom:5pt"></div>

The **Euler method** uses the following iteration to find values of $y$:
<div style="margin-top:5pt"></div>
$\large y_{k+1}=y_k+f(y_k,t_k)\cdot \Delta t, k=0,1,2,...$
<div style="margin-bottom:5pt"></div>
where $t_k=t_0+k\cdot \Delta t$, and $y_k:=y(t_k)$. Also, $\Delta t$ is the step size along t-axis, which is a small positive number. 
<hr>

In the following, we implement the Euler method for the IVP: $\dot{y}=-y$ and $y(0)=1$

```python
# Import required modules
from math import exp
from matplotlib import pyplot as plt
```

```python
# Define Euler method for y'=f(y,t)
def euler_method(f_yt,y0,t0=0,dt=0.1, steps=100):
    y,t = y0,t0
    trajectory = [y0]
    ts=[t0]
    for _ in range(steps):
        y += f_yt(y,t) * dt  # Euler update
        trajectory.append(y)
        t+=dt
        ts.append(t)
    return ts, trajectory
```

```python
# Define the ODE function as: dy/dt = -y (exponential decay)
def dy_dt(y,t):
    return -y
```

```python
# Simulate with Î”t = 0.1
ts,ys = euler_method(dy_dt,y0=1.0,t0=0)
ys_true=[exp(-t) for t in ts] # true solution: y(t)=exp(-t)
plt.plot(ts,ys,label='Euler method',c='blue')
plt.plot(ts,ys_true,c='red', label='True solution')
plt.title('Euler method versus true solution for $\dot{y}=-y,y(0)=1$')
plt.legend(); plt.grid()
plt.xlabel('$t$'); plt.ylabel('$y$')
plt.show()
```
