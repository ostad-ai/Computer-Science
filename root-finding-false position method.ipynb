# Root finding with False Position method
Consider the equation $f(x)=0$ for real variable $x$ where $f$ is continuous on the interval $[x_L,x_R]$.
and $f(x_L)f(x_R)<0$; then, there exists at least one root of *f* in the interval. 
<br> The numbers $x_L$ and $x_R$ are given as they bracket the root. **False Position method** is a kind of **bracketing method** which can be used to extract one root from the given interval $[x_L,x_R]$.
<br>**Hint:** We saw earlier the bisection method, which is also a bracketing method. False position method includes the magnitude of function at the two endpoints, which makes it a better method.

```python
# importing modules
import numpy as np
from matplotlib import pyplot as plt
```

```python
# defining the function we seek its root
def funct(x):
    return np.cos(x/2.)
#--- display the function in the interval
left,right=0,8
xs=np.linspace(left,right,100)
plt.plot(xs,funct(xs))
plt.axhline(color='red',linestyle='-.')
plt.title('$f(x)=cos(x/2),\quad x\in[{0},{1}]$'\
          .format(left,right))
plt.xlabel('x')
plt.show()
```

```python
# the false position method from scratch
def FalsePosition(f,xL,xR,maxiter=100,tol=1e-15):
    if (f(xL)*f(xR))>0 or (xL>xR):
        return None,None #error in parameters
    side='none'
    fL=f(xL); fR=f(xR); xOld=xL
    for _ in range(maxiter):
        xC=xR-fR*(xL-xR)/(fL-fR)
        if xC!=0:
            e_actual=(xC-xOld)/xC  
        else:
            e_actual=XC-xOld
        fC=f(xC)
        if abs(e_actual)<tol:
            break
        test=fL*fC
        if test<0:
            xR=xC; fR=fC
            if side=='left':
                fL/=2.
            side='left'
        elif test>0:
            xL=xC; fL=fC
            if side=='right':
                fR/=2.
            side='right'
        else:
            break
        xOld=xC
    return xC,fC
```

```python
# testing the false position method from scratch
x,fx=FalsePosition(funct,0,8)
print(f'root: {x}\nvalue of function at root: {fx}')
print(f'value of PI in numpy:{np.pi}')
```
